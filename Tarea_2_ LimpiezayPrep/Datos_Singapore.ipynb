````python
import re

def parse_bathroom_info(bathrooms_text):
    if not isinstance(bathrooms_text, str):
        return np.nan, np.nan

    text = bathrooms_text.lower()
    is_shared = 'shared' in text
    num_bathrooms = np.nan

    # Handle 'half-bath' cases
    if 'half-bath' in text or 'half bath' in text:
        if 'private half-bath' in text or 'shared half-bath' in text:
            num_bathrooms = 0.5
        elif 'half-bath' in text:
            num_bathrooms = 0.5
        else:
            # Try to extract number before 'half-bath' if any, e.g., '1.5 shared baths'
            match = re.search(r'(\d+\.?\d*)\s*(half)?\s*baths?', text)
            if match:
                num_bathrooms = float(match.group(1)) if match.group(1) else np.nan
                if 'half' in text and num_bathrooms is not np.nan:
                    num_bathrooms *= 0.5 # This case handles like 1.5 baths, not '1 half-bath'

    # Extract numerical value directly
    if num_bathrooms is np.nan:
        match = re.search(r'(\d+\.?\d*)', text)
        if match:
            num_bathrooms = float(match.group(1))
        elif 'bath' in text and not match and '0' in text: # handles '0 baths'
            num_bathrooms = 0.0

    # Special handling for 'private half-bath' or 'shared half-bath'
    if 'private half-bath' in text and not is_shared:
        num_bathrooms = 0.5
    elif 'shared half-bath' in text and is_shared:
        num_bathrooms = 0.5

    return num_bathrooms, is_shared
````

````python
bathroom_info = data['bathrooms_text'].apply(parse_bathroom_info)
data['number_of_bathrooms_temp'] = [x[0] for x in bathroom_info]
data['is_shared_temp'] = [x[1] for x in bathroom_info]

# Populate the new columns based on 'is_shared_temp'
data.loc[data['is_shared_temp'] == False, 'number_of_private_bathrooms'] = data['number_of_bathrooms_temp']
data.loc[data['is_shared_temp'] == True, 'number_of_shared_bathrooms'] = data['number_of_bathrooms_temp']

# Drop the temporary columns
data.drop(columns=['number_of_bathrooms_temp', 'is_shared_temp'], inplace=True)
````

````python
# Sum the values from 'number_of_private_bathrooms' and 'number_of_shared_bathrooms'
# Use .add() with fill_value=0 to treat NaN values as 0 during summation
data['bathrooms'] = data['number_of_private_bathrooms'].add(data['number_of_shared_bathrooms'], fill_value=0)
````

````python
data['bedrooms'] = data['bedrooms'].replace(0, np.nan)
data['bedrooms'] = data['bedrooms'].fillna(data['bathrooms'])
data['bedrooms'] = np.ceil(data['bedrooms'])
````

````python
average_bedrooms_by_room_type = data.groupby('room_type')['bedrooms'].mean().apply(np.ceil)
print("Average bedrooms per room_type (rounded up):")
display(average_bedrooms_by_room_type)
# Identify rows where 'bedrooms' is 0.0
zero_bedrooms_mask = data['bedrooms'] == 0.0

# Loop through each room_type and replace 0.0 bedrooms with the calculated average
for room_type, avg_bedrooms in average_bedrooms_by_room_type.items():
    data.loc[zero_bedrooms_mask & (data['room_type'] == room_type), 'bedrooms'] = avg_bedrooms
````

````python
average_beds_by_bedrooms = data.groupby('bedrooms')['beds'].mean().apply(np.ceil)
print("Average beds per bedrooms (rounded up):")
display(average_beds_by_bedrooms)
# Identify rows where 'beds' is either NaN or 0.0
beds_to_impute_mask = data['beds'].isnull() | (data['beds'] == 0.0)

# Loop through each 'bedrooms' category and its calculated average 'beds'
for bedrooms_count, avg_beds in average_beds_by_bedrooms.items():
    # Apply imputation only for rows where 'beds' needs to be imputed and 'bedrooms' matches
    data.loc[beds_to_impute_mask & (data['bedrooms'] == bedrooms_count), 'beds'] = avg_beds
````

````python
response_time_mapping = {
    'within an hour': 1,
    'within a day': 24,
    'a few days or more': 48,
    'within a few hours': 12
}

data['estimated_response_time_hours'] = data['host_response_time'].map(response_time_mapping)

# Calculate the mean 'estimated_response_time_hours' for each 'host_id'
host_response_time_hours_avg = data.groupby('host_id')['estimated_response_time_hours'].transform('mean')

# Fill missing values in 'estimated_response_time_hours' with the calculated averages
data['estimated_response_time_hours'].fillna(host_response_time_hours_avg, inplace=True)

# Fill any remaining missing 'estimated_response_time_hours' values with the overall mean
overall_mean_estimated_response_time = data['estimated_response_time_hours'].mean()
data['estimated_response_time_hours'].fillna(overall_mean_estimated_response_time, inplace=True)
````

````python
data['host_response_rate'] = data['host_response_rate'].astype(str).str.replace('%', '', regex=False)
data['host_response_rate'] = pd.to_numeric(data['host_response_rate'], errors='coerce') / 100

# Calculate the mean 'host_response_rate' for each 'host_id'
host_response_rate_avg = data.groupby('host_id')['host_response_rate'].transform('mean')

# Fill missing values in 'host_response_rate' with the calculated averages
data['host_response_rate'] = data['host_response_rate'].fillna(host_response_rate_avg)

# Fill any remaining missing 'host_response_rate' values with the overall mean
overall_mean_host_response_rate = data['host_response_rate'].mean()
data['host_response_rate'] = data['host_response_rate'].fillna(overall_mean_host_response_rate)
````

````python
communication_score_mapping = {
    'within an hour': 4,
    'within a few hours': 3.5,
    'within a day': 3,
    'a few days or more': 2.5
}

# Convert 'host_response_time' to numerical score
data['host_response_time_score'] = data['host_response_time'].map(communication_score_mapping)

# Calculate the combined score
data['combined_communication_score'] = data['host_response_time_score'] + data['host_response_rate']

# Fill missing values in 'review_scores_communication' with the combined score
data['review_scores_communication'].fillna(data['combined_communication_score'], inplace=True)

del data['host_response_time_score']
del data['combined_communication_score']

print("Columnas temporales eliminadas.")
del data['host_response_time']
````

````python
# Ensure 'host_acceptance_rate' is numeric before calculating means
data['host_acceptance_rate'] = data['host_acceptance_rate'].str.replace('%', '', regex=False).astype(float) / 100

# Calculate the mean 'host_acceptance_rate' for each 'host_id'
host_acceptance_rate_avg = data.groupby('host_id')['host_acceptance_rate'].transform('mean')

# Fill missing values in 'host_acceptance_rate' with the calculated averages
data['host_acceptance_rate'].fillna(host_acceptance_rate_avg, inplace=True)

# Fill any remaining missing 'host_acceptance_rate' values with the overall mean
overall_mean_acceptance_rate = data['host_acceptance_rate'].mean()
data['host_acceptance_rate'].fillna(overall_mean_acceptance_rate, inplace=True)
````
